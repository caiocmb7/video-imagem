```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
imagem_cv2 = cv2.imread('/content/Galactic_Silhouettes.jpg') # Ler imagem
imagem_cv2 = cv2.resize(imagem_cv2, (200,200))
imgGray = cv2.cvtColor(imagem_cv2, cv2.COLOR_BGR2GRAY)
fig = plt.figure(figsize=(10, 6))
plt.subplot(121); plt.axis("off"); plt.imshow(imagem_cv2)
plt.subplot(122); plt.axis("off"), plt.imshow(imgGray, cmap = 'gray');
```

```python
cv2.imwrite('/content/Galactic_Silhouettes_Gray.jpg', imgGray) # salvar a img cinza
```

```python
# Filtro da mdia
blur = cv2.blur(imgGray,(5,5))
blur2 = cv2.blur(imgGray,(10,10))
blur3 = cv2.blur(imgGray,(20,20))
blur4 = cv2.blur(imgGray,(50,50))
blur5 = cv2.blur(imgGray,(100,100))
fig = plt.figure(figsize=(20, 6))
plt.subplot(151);plt.axis("off"); plt.imshow(blur,cmap='gray')
plt.subplot(152);plt.axis("off"); plt.imshow(blur2,cmap='gray')
plt.subplot(153);plt.axis("off"); plt.imshow(blur3,cmap='gray')
plt.subplot(154);plt.axis("off"); plt.imshow(blur4,cmap='gray')
plt.subplot(155);plt.axis("off"); plt.imshow(blur5,cmap='gray')
plt.title("Filtro da mdia")
plt.show()
```

```python
imagem = Image.open("/content/Galactic_Silhouettes_Gray.jpg")
imagem
```

```python
img_filter = imagem.filter(ImageFilter.GaussianBlur(3))

fig = plt.figure(figsize=(8,6))
plt.subplot(121); plt.axis("off"); plt.imshow(imagem)
plt.subplot(122); plt.axis("off"); plt.imshow(img_filter)
```
